```python
import pandas as pd
import numpy as np
from faker import Faker
import os
```

```python
# Initialize Faker
fake = Faker()

# Define the base directory
base_dir = '/home/joserico/Desktop/FLServer/scenarios/technological'

# Function to generate IID datasets with equal row count and normal distribution
def generate_iid_normal_dataset(num_rows, mean=50, std_dev=10):
    target_variable = np.random.choice([0, 1], size=num_rows)
    feature_columns = {
        f'Feature_{i}': np.random.normal(loc=mean, scale=std_dev, size=num_rows)
        for i in range(1, 8)
    }
    df = pd.DataFrame({
        'TargetVariable': target_variable,
        **feature_columns
    })
    return df

# Function to generate exaggerated non-IID datasets with different row counts
def generate_non_iid_dataset(num_rows, imbalance_ratio=0.7):
    num_class_0 = int(num_rows * imbalance_ratio)
    num_class_1 = num_rows - num_class_0
    target_variable = np.array([0] * num_class_0 + [1] * num_class_1)
    np.random.shuffle(target_variable)
    
    # Use Faker to create more variability in features
    feature_columns = {
        f'Feature_{i}': [fake.random_number(digits=3) * (1 if i % 2 == 0 else 10) + np.random.rand() for _ in range(num_rows)]
        for i in range(1, 8)
    }
    
    df = pd.DataFrame({
        'TargetVariable': target_variable,
        **feature_columns
    })
    return df
```

```python
# Generate and save datasets for each client
for i in range(1, 6):
    # Generate IID dataset with normal distribution and fixed row count
    df_iid = generate_iid_normal_dataset(5000)
    file_name_iid = f'{base_dir}/client{i}/data{i}_iid.csv'
    df_iid.to_csv(file_name_iid, index=False)
    
    # Generate exaggerated non-IID dataset with variable row count
    non_iid_rows = np.random.randint(3000, 7000)  # Random row count for non-IID datasets
    imbalance_ratio = np.random.uniform(0.1, 0.9)  # Random imbalance ratio
    df_non_iid = generate_non_iid_dataset(non_iid_rows, imbalance_ratio=imbalance_ratio)
    file_name_non_iid = f'{base_dir}/client{i}/data{i}_nonIID.csv'
    df_non_iid.to_csv(file_name_non_iid, index=False)

print("IID and non-IID datasets generated and saved successfully!")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define paths for IID and non-IID datasets
iid_dataset_path = '/home/joserico/Desktop/FLServer/scenarios/technological/client1/data1_iid.csv'  # Replace with your path
non_iid_dataset_path = '/home/joserico/Desktop/FLServer/scenarios/technological/client1/data1_nonIID.csv'  # Replace with your path

# Load the datasets
df_iid = pd.read_csv(iid_dataset_path)
df_non_iid = pd.read_csv(non_iid_dataset_path)

def analyze_distribution(df, dataset_name):
    print(f"Analyzing {dataset_name} dataset...")
    
    # 1. Histogram for each feature
    for feature in df.columns[:-1]:  # Exclude the 'TargetVariable'
        plt.figure(figsize=(8, 6))
        sns.histplot(df[feature], kde=True)
        plt.title(f'Histogram of {feature} in {dataset_name}')
        plt.xlabel(feature)
        plt.ylabel('Frequency')
        plt.show()

    # 2. Q-Q Plot for each feature
    for feature in df.columns[:-1]:
        plt.figure(figsize=(8, 6))
        stats.probplot(df[feature], dist="norm", plot=plt)
        plt.title(f'Q-Q Plot of {feature} in {dataset_name}')
        plt.show()

    # 3. Shapiro-Wilk Test for Normality
    for feature in df.columns[:-1]:
        stat, p = stats.shapiro(df[feature])
        print(f'Shapiro-Wilk Test for {feature} in {dataset_name}:')
        print(f'Statistic: {stat}, p-value: {p}')
        if p > 0.05:
            print(f'{feature} follows a normal distribution (fail to reject H0)')
        else:
            print(f'{feature} does not follow a normal distribution (reject H0)')
        print('\n')

# Analyze IID dataset
analyze_distribution(df_iid, "IID")

# Analyze non-IID dataset
analyze_distribution(df_non_iid, "Non-IID")
```
