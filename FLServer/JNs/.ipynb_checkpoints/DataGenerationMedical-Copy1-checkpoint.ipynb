```python
# Importing required packages
import os # version Python: 3.10.12
# Set environment variables to suppress specific TensorFlow messages
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
import warnings # Standard Library
warnings.filterwarnings("ignore", category=UserWarning)
import tensorflow as tf # version 2.14.1
import logging # Standard Library
logging.basicConfig(level=logging.INFO)
import cv2 # version 4.10.0
import zipfile # Standard Library
import requests # version 2.32.3
import numpy as np # version 1.25.2
import matplotlib.pyplot as plt # version 1.25.2
from tensorflow.keras.datasets import mnist # version 2.14.1
from tensorflow.keras.preprocessing.image import ImageDataGenerator # version 2.14.1
```

```python
# Define paths for downloading and storing the ChestXRay2017 dataset
base_path = '/home/joserico/Desktop/Capstone_MScData_Sept23_SB/FLServer/scenarios/medical' # Adjust this path accordignly
zip_file_name = 'ChestXRay2017.zip' # Removing this folder after file creation, keeping GitHub repo as light as possible                                        
zip_file_path = os.path.join(base_path, zip_file_name)
extracted_dir_path = os.path.join(base_path, 'ChestXRay2017') # Removing this folder after file creation, keeping GitHub repo as light as possible 

# URL to download the ChestXRay2017 dataset
download_url = "https://data.mendeley.com/public-files/datasets/rscbjbr9sj/files/f12eaf6d-6023-432f-acc9-80c9d7393433/file_downloaded"

# Function to download the dataset if it doesn't exist locally
def download_dataset(url, save_path):
    if not os.path.exists(save_path):
        print(f"Downloading dataset from {url}...")
        response = requests.get(url, stream=True)
        with open(save_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=128):
                file.write(chunk)
        print(f"Dataset downloaded and saved to {save_path}.")
    else:
        print(f"Dataset already downloaded at {save_path}.")

# Function to extract the dataset from the downloaded zip file
def extract_dataset(zip_file, extract_path):
    if not os.path.exists(extract_path):
        with zipfile.ZipFile(zip_file, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
        print(f"Dataset extracted to {extract_path}.")
    else:
        print(f"Dataset already extracted at {extract_path}.")

# Function to load and preprocess the MNIST dataset (resize and convert to RGB)
def load_and_preprocess_mnist():
    (x_train, y_train), (_, _) = mnist.load_data()
    x_train = np.stack([cv2.resize(img, (64, 64)) for img in x_train])
    x_train = np.stack([cv2.cvtColor(img, cv2.COLOR_GRAY2RGB) for img in x_train])
    y_train = np.zeros(len(y_train))  # Label MNIST as 0 (NotLung)
    return x_train, y_train

# Function to load and preprocess the Chest X-ray dataset (resize and normalize)
def load_pneumonia_dataset(base_dir):
    datagen = ImageDataGenerator(rescale=1./255)
    dataset = datagen.flow_from_directory(
        base_dir,
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary',
        shuffle=True
    )
    
    images, labels = [], []
    for i in range(len(dataset)):
        images.extend(dataset[i][0])
        labels.extend(dataset[i][1])
    images = np.array(images)
    labels = np.array(labels)
    labels = np.ones(len(labels))  # Label all X-ray images as 1 (Lung)
    return images, labels

# Function to generate image names in a sequence like A, B, ..., Z, AA, AB, etc.
def generate_image_name(index):
    name = ""
    while index >= 0:
        name = chr((index % 26) + 65) + name
        index = index // 26 - 1
    return name

# Function to save images into the specified directory structure for each client
def save_images_to_directory(images, labels, base_dir, client_id, setting, is_iid=True, dist=None):
    # Define the number of training and test images to save for each scenario
    if is_iid:
        max_train = 150  # Reduced number of images for IID scenario
        max_test = 150
    else:
        max_train = dist['train']
        max_test = dist['test']

    saved_train_count = 0
    saved_test_count = 0

    # Iterate through both classes (NotLung and Lung) and save images accordingly
    for label in [0, 1]:
        class_images = images[labels == label]
        label_dir = 'Lung' if label == 1 else 'NotLung'
        
        for i, image in enumerate(class_images):
            if i < max_train:
                subset = 'train'
                saved_train_count += 1
            elif i < max_train + max_test:
                subset = 'test'
                saved_test_count += 1
            else:
                break

            save_dir = os.path.join(base_dir, f'client{client_id}', setting, subset, label_dir)
            os.makedirs(save_dir, exist_ok=True)

            # Generate file names based on class: numeric for Lung, alphabetical for NotLung
            if label == 1:
                image_name = f'{i + 1}.jpg'
            else:
                image_name = f'{generate_image_name(i)}.jpg'

            image_path = os.path.join(save_dir, image_name)
            cv2.imwrite(image_path, image * 255)  # Convert normalized images back to 0-255 range
        
        # Only print a message if images were saved
        if saved_train_count > 0 or saved_test_count > 0:
            print(f"Saved {saved_train_count} training images and {saved_test_count} test images in {label_dir} for client {client_id} in {setting}.")

# Function to generate and save IID datasets for multiple clients
def generate_and_save_iid_datasets(mnist_data, xray_data, client_base_path, num_clients=5):
    mnist_images, mnist_labels = mnist_data
    xray_images, xray_labels = xray_data
    
    for client_id in range(1, num_clients + 1):
        print(f"Saving IID dataset for Client {client_id}...")
        save_images_to_directory(mnist_images, mnist_labels, client_base_path, client_id, 'IID', is_iid=True)
        save_images_to_directory(xray_images, xray_labels, client_base_path, client_id, 'IID', is_iid=True)
        print(f"Dataset for Client {client_id} IID images saved successfully.")

# Function to generate and save non-IID datasets for multiple clients
def generate_and_save_non_iid_datasets(mnist_data, xray_data, client_base_path, num_clients=5):
    mnist_images, mnist_labels = mnist_data
    xray_images, xray_labels = xray_data
    
    # Define exaggerated distributions for non-IID scenario
    client_distributions = [
        {'NotLung': {'train': 10, 'test': 60}, 'Lung': {'train': 100, 'test': 50}},  # Client 1
        {'NotLung': {'train': 15, 'test': 100}, 'Lung': {'train': 40, 'test': 50}},  # Client 2
        {'NotLung': {'train': 80, 'test': 30}, 'Lung': {'train': 50, 'test': 40}},  # Client 3
        {'NotLung': {'train': 20, 'test': 90}, 'Lung': {'train': 35, 'test': 100}},  # Client 4
        {'NotLung': {'train': 80, 'test': 30}, 'Lung': {'train': 100, 'test': 40}},  # Client 5
    ]
    
    for client_id in range(1, num_clients + 1):
        dist = client_distributions[client_id - 1]
        print(f"Saving non-IID dataset for Client {client_id}...")
        
        save_images_to_directory(mnist_images, mnist_labels, client_base_path, client_id, 'nonIID', is_iid=False, dist=dist['NotLung'])
        save_images_to_directory(xray_images, xray_labels, client_base_path, client_id, 'nonIID', is_iid=False, dist=dist['Lung'])
        print(f"Dataset for Client {client_id} non-IID images saved successfully.")

# Download the ChestXRay2017 dataset from the given URL
download_dataset(download_url, zip_file_path)

# Extract the downloaded dataset
extract_dataset(zip_file_path, extracted_dir_path)

# Load and preprocess MNIST and Chest X-ray datasets
mnist_images_iid, mnist_labels_iid = load_and_preprocess_mnist()
xray_images_iid, xray_labels_iid = load_pneumonia_dataset(os.path.join(extracted_dir_path, 'chest_xray', 'train'))

# Create copies of the data for non-IID scenario
mnist_images_non_iid, mnist_labels_non_iid = mnist_images_iid.copy(), mnist_labels_iid.copy()
xray_images_non_iid, xray_labels_non_iid = xray_images_iid.copy(), xray_labels_iid.copy()

# Generate and save IID datasets for all clients
generate_and_save_iid_datasets((mnist_images_iid, mnist_labels_iid), (xray_images_iid, xray_labels_iid), base_path)

# Generate and save non-IID datasets for all clients
generate_and_save_non_iid_datasets((mnist_images_non_iid, mnist_labels_non_iid), (xray_images_non_iid, xray_labels_non_iid), base_path)

# Function to count the number of images in a directory (including subdirectories)
def count_images_in_directory(directory):
    count = 0
    for root, dirs, files in os.walk(directory):
        count += len([f for f in files if f.endswith('.jpg') or f.endswith('.png')])
    return count
```

```python

# Function to plot pie charts for image distributions with a common legend
def plot_pie_charts_with_legend(base_path):
    settings = ['IID', 'nonIID']
    labels = ['Lung - Train', 'Lung - Test', 'NotLung - Train', 'NotLung - Test']
    
    # Monochromatic color scheme: various shades of gray-green
    colors = ['#4d4d4d', '#6a6a6a', '#8c8c8c', '#b3b3b3']  # Adjust shades as needed

    fig, axes = plt.subplots(2, 5, figsize=(20, 10))  # Adjusted to fit both IID and non-IID settings

    for setting_idx, setting in enumerate(settings):
        for client_id in range(1, 6):  # Assuming 5 clients
            lung_train_dir = os.path.join(base_path, f'client{client_id}', setting, 'train', 'Lung')
            lung_test_dir = os.path.join(base_path, f'client{client_id}', setting, 'test', 'Lung')
            notlung_train_dir = os.path.join(base_path, f'client{client_id}', setting, 'train', 'NotLung')
            notlung_test_dir = os.path.join(base_path, f'client{client_id}', setting, 'test', 'NotLung')

            counts = [
                count_images_in_directory(lung_train_dir),
                count_images_in_directory(lung_test_dir),
                count_images_in_directory(notlung_train_dir),
                count_images_in_directory(notlung_test_dir)
            ]

            ax = axes[setting_idx, client_id - 1]
            ax.pie(counts, colors=colors, autopct=lambda p: '{:.1f}%'.format(p) if p > 0 else '', 
                   startangle=90, textprops={'fontsize': 16})
            ax.set_title(f'Client {client_id} ({setting})', fontsize=16)

    # Adjust the space between plots and title
    fig.subplots_adjust(hspace=0.2, wspace=0.3, top=0.85)

    # Add a common legend at the bottom of the plot area
    fig.legend(labels, loc='lower center', bbox_to_anchor=(0.5, 0.009), ncol=4, fontsize=16, frameon=False)
    plt.suptitle('Image Distribution Across Clients (IID and nonIID)', fontsize=20)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to make room for the legend
    plt.show()

# Define the base path for client datasets
base_path = '/home/joserico/Desktop/FLServer/scenarios/medical'

# Plot pie charts with a common legend for image distribution across clients
plot_pie_charts_with_legend(base_path)
```

# Summary of Key Functions

## 1. `download_dataset(url, save_path)`

**Purpose:**  
Downloads the ChestXRay2017 dataset from the provided URL if it doesn't already exist locally.

**Inputs:**
- `url`: The URL to download the dataset from.
- `save_path`: The local path to save the downloaded file.

**Output:**  
A downloaded dataset file at the specified location.

---

## 2. `extract_dataset(zip_file, extract_path)`

**Purpose:**  
Extracts the downloaded dataset from the zip file.

**Inputs:**
- `zip_file`: Path to the zip file.
- `extract_path`: Directory to extract the contents of the zip file.

**Output:**  
Extracted dataset files in the specified directory.

---

## 3. `load_and_preprocess_mnist()`

**Purpose:**  
Loads and preprocesses the MNIST dataset by resizing the images to 64x64 pixels and converting grayscale images to RGB.

**Inputs:**  
None.

**Output:**  
A tuple containing the processed images and labels, with MNIST images labeled as NotLung.

---

## 4. `load_pneumonia_dataset(base_dir)`

**Purpose:**  
Loads and preprocesses the Chest X-ray dataset, resizing images to 64x64 pixels and normalizing pixel values.

**Inputs:**
- `base_dir`: The base directory where the Chest X-ray dataset is located.

**Output:**  
A tuple containing the processed images and labels, with all X-ray images labeled as Lung.

---

## 5. `generate_image_name(index)`

**Purpose:**  
Generates image filenames in a sequence (e.g., A, B, ..., Z, AA, AB, etc.).

**Inputs:**
- `index`: The index number used to generate the filename.

**Output:**  
A string representing the generated image filename.

---

## 6. `save_images_to_directory(images, labels, base_dir, client_id, setting, is_iid=True, dist=None)`

**Purpose:**  
Saves the images into the specified directory structure for each client, creating train and test subsets.

**Inputs:**
- `images`: Array of image data.
- `labels`: Array of image labels.
- `base_dir`: The base directory for saving images.
- `client_id`: The client identifier (1 to 5).
- `setting`: Either 'IID' or 'nonIID' to specify the dataset setting.
- `is_iid`: Boolean flag indicating if the dataset is IID or non-IID.
- `dist`: Dictionary specifying the number of train and test images for the non-IID setting.

**Output:**  
Saved image files organized by client, setting, and class directories.

---

## 7. `generate_and_save_iid_datasets(mnist_data, xray_data, client_base_path, num_clients=5)`

**Purpose:**  
Generates and saves IID datasets for multiple clients.

**Inputs:**
- `mnist_data`: Tuple containing MNIST images and labels.
- `xray_data`: Tuple containing X-ray images and labels.
- `client_base_path`: The base directory for saving client datasets.
- `num_clients`: Number of clients to generate datasets for.

**Output:**  
IID datasets saved for each client.

---

## 8. `generate_and_save_non_iid_datasets(mnist_data, xray_data, client_base_path, num_clients=5)`

**Purpose:**  
Generates and saves non-IID datasets for multiple clients with exaggerated data distribution.

**Inputs:**  
Similar to `generate_and_save_iid_datasets`, with additional distribution settings for non-IID data.

**Output:**  
Non-IID datasets saved for each client.

---

## 9. `count_images_in_directory(directory)`

**Purpose:**  
Counts the number of image files in a specified directory, including its subdirectories.

**Inputs:**
- `directory`: The directory path to count image files.

**Output:**  
An integer representing the total number of image files found.

---

## 10. `plot_pie_charts_with_legend(base_path)`

**Purpose:**  
Plots pie charts for image distributions across clients in IID and non-IID settings, with a common legend.

**Inputs:**
- `base_path`: The base directory where client datasets are stored.

**Output:**  
Displayed pie charts showing the distribution of images across clients and settings.
